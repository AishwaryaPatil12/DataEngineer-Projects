###Create Schemas

```python
%sql
use catalog customersdata;
-- create schemas in catalog customersdata
create schema if not exists bronze;
create schema if not exists silver;
create schema if not exists gold;

--use bronze schema 
use schema bronze;
```

###Load data from csv files

```python
from pyspark.sql import functions as F

#Read data from customers file
df_customers_bronze = spark.read.csv("/Volumes/customersdata/bronze/rawfiles/customers.csv", header=True, inferSchema=True)
display(df_customers_bronze)

#Read data from orders file
df_orders_bronze = spark.read.csv("/Volumes/customersdata/bronze/rawfiles/orders.csv", header=True, inferSchema=True)
display(df_orders_bronze)
```

###Save raw data into bronze tables

```python
#Save customers data in bronze table
df_customers_bronze.write.mode("overwrite").saveAsTable("customersdata.bronze.customers_bronze")

#Save orders data in bronze table
df_orders_bronze.write.mode("overwrite").saveAsTable("customersdata.bronze.orders_bronze")
```

```python
%sql
select * from customersdata.bronze.customers_bronze;

select * from customersdata.bronze.orders_bronze;
```
